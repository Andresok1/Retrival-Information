Solution:

E1 = - (0.5+0.5)* 3 * (1/3) * log2(1/3) = 1.585

E2 = - (2/3) * (2* (1/2) * log2(1/2)) - (1/3) * (1 * log2(1)) = 0.666

G1 = E - E1 = 0
