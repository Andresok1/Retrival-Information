### Question 1:

Consider the document collection given below. Create an inverted index from it and answer the following queries:

- brutus AND caesar
- friends AND romans AND countrymen

Hint: Please do a case folding, and remove punctuation marks. You can also remove genitive s (not required), but any other preprocessing is not needed for this task.

```python
#Document collection
d1 = 'Et tu, Brute! Then fall, Caesar'
d2 = "Friends, Romans, countrymen, lend me your ears; I come to bury Caesar, not to praise him. Come I to speak in Caesar's funeral. He was my friend, faithful and just to me: But Brutus says he was ambitious; And Brutus is an honourable man."
d3 = "Nor heaven nor earth have been at peace to-night: Thrice hath Calpurnia in her sleep cried out, 'Help, ho! they murder Caesar!' Who's within?"
d4 = "Romans, countrymen, and lovers! hear me for my cause, and be silent, that you may hear: If there be any in this assembly, any dear friend of Caesar's, to him I say, that Brutus' love to Caesar was no less than his. If then that friend demand why Brutus rose against Caesar, this is my answer: Not that I loved Caesar less, but that I loved Rome more."
```

```python
#your code comes here:
```

```python
#one possible solution
d1 = d1.split()
d1_words =[]
for word in d1:
    d1_words.append(word.strip("',.!;:?").replace("'s", '').lower())
print(d1_words)

d2 = d2.split()
d2_words =[]
for word in d2:
    d2_words.append(word.strip("',.!;:?").replace("'s", '').lower())
print(d2_words)

d3 = d3.split()
d3_words =[]
for word in d3:
    d3_words.append(word.strip("',.!;:?").replace("'s", '').lower())
print(d3_words)

d4 = d4.split()
d4_words =[]
for word in d4:
    d4_words.append(word.strip("',.!;:?").replace("'s", '').lower())
print(d4_words)
```

```python
docCollection = [d1_words, d2_words, d3_words, d4_words]
invIndex = {}
for i, doc in enumerate(docCollection):
    for word in doc:
        if word in invIndex.keys():
            if not i in invIndex[word]:
                invIndex[word].append(i)
        else:
            invIndex[word]=[i]
invIndex
invIndex2 = dict(sorted(invIndex.items()))
invIndex2
```

```python
def intersect(p1, p2):
    answer = []
    i = 0
    j = 0
    while i < len(p1) and j < len(p2):
        if p1[i] == p2[j]:
            answer.append(p1[i])
            i+=1
            j+=1
        else:
            if p1[i] < p2[j]:
                i += 1
            else:
                j += 1
    return answer
```

```python
print(intersect(invIndex['brutus'], invIndex['caesar']))
```

```python
print(intersect(invIndex['friends'], intersect(invIndex['romans'], invIndex['countrymen'])))
```

### Question 2:

Change your code to solve the following queries:  

- caesar AND NOT brutus
- friends OR lovers

```python
#possible answer

#OR:

def intersect_or(p1, p2):
    answer = []
    i = 0
    j = 0
    while i < len(p1) and j < len(p2):
        answer.append(p1[i])
        answer.append(p2[j])
        if i < len(p1)-1 and j < len(p2)-1:
            i+=1
            j+=1
        else:
            if j == len(p2)-1 and i < len(p1)-1:
                i+=1
            elif i == len(p1)-1 and j < len(p2)-1:
                j+=1
            else:
                i+=1
                j+=1
    answer = sorted(set(answer))
    return answer

#add any document ID in p1 or p2 to the answer, if it is not in the answer already.

#AND NOT:

def intersect_andnot(p1, p2):
    answer = []
    i = 0
    j = 0
    while i < len(p1) and j < len(p2):
        if p1[i] == p2[j]: # if it's identical, go one step up for both
            i+=1
            j+=1
        elif p1[i] < p2[j]: # if the document ID of the second term is larger, than add the doc ID of the first term to the answer
            answer.append(p1[i])
            i +=1
        elif p1[i] > p2[j]: # if the document ID of the second term is smaller, move it up, until it is same or larger
            j+=1
    return answer

```

```python
print(intersect_or(invIndex['friends'], invIndex['lovers']))
```

```python
print(intersect_andnot(invIndex['caesar'], invIndex['brutus']))
```
