```python
import pyterrier as pt
import pyterrier_dr
```

```python
model = pyterrier_dr.TctColBert()

# Loads castorini/tct_colbert-msmarco by default.
# You can load up other versions by specifying the huggingface model ID, e.g.,
model = pyterrier_dr.TctColBert('castorini/tct_colbert-v2-hnp-msmarco')

model = pyterrier_dr.TasB()

print(model)
```

```python
# Compute query vectors
output = model([
  {'qid': '0', 'query': 'Hello Terrier'},
  {'qid': '1', 'query': 'find me some documents'},
])

print(output)

print("*"*100)

# Compute document vectors
output =  model([
  {'docno': '0', 'text': 'The Five Find-Outers and Dog, also known as The Five Find-Outers, is a series of children\'s mystery books written by Enid Blyton.'},
  {'docno': '1', 'text': 'City is a 1952 science fiction fix-up novel by American writer Clifford D. Simak.'},
])

print(output)

print("*"*100)
# Compute on-they-fly scores
output = model([
  {'qid': '0', 'query': 'Hello Terrier', 'docno': '0', 'text': 'The Five Find-Outers and Dog, also known as The Five Find-Outers, is a series of children\'s mystery books written by Enid Blyton.'},
  {'qid': '0', 'query': 'Hello Terrier', 'docno': '1', 'text': 'City is a 1952 science fiction fix-up novel by American writer Clifford D. Simak.'},
])

print(output)

print("*"*100)
```

BM25 Retriever Sparse Index
Total size~3.20GB, Time to download~2 mins 

You will need to install java for Terrier  `sudo apt-get update && sudo apt-get install -y default-jdk`


```python
# It will download the sparse index from HF Hub

sparse_index = pt.Artifact.from_hf('macavaney/msmarco-passage.terrier')

print(sparse_index)

print(sparse_index.path)

sparse_index = pt.IndexFactory.of(f"{sparse_index.path}")

bm25 = pt.terrier.Retriever(sparse_index, wmodel="BM25")
```

```python
bm25.search('Introduction to Information Retrieval')
```

```python
from pyterrier.measures import *
dataset = pt.get_dataset('irds:msmarco-passage/trec-dl-2019/judged')
pt.Experiment(
    [bm25 % 100,
    retr_pipeline % 100],
    dataset.get_topics(),
    dataset.get_qrels(),
    [nDCG, MAP(rel=2), R(rel=2)@1000],
    names=['BM25', 'TaSB']
)
```

With Qwen3 Model  https://huggingface.co/Qwen/Qwen3-Embedding-0.6B 

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
```

```python
queries = [
    "What is the capital of China?",
    "Explain gravity",
]
documents = [
    "The capital of China is Beijing.",
    "Gravity is a force that attracts two bodies towards each other. It gives weight to physical objects and is responsible for the movement of planets around the sun.",
]
```

```python
query_embeddings = model.encode(queries, prompt_name="query")
document_embeddings = model.encode(documents)

# Compute the (cosine) similarity between the query and document embeddings
similarity = model.similarity(query_embeddings, document_embeddings)
print(similarity)
```
