# Tutorial 9 and 10: Link Analysis and Crawling

## Link Analysis

1. State two advantages of the Boldi and Vigna (2004) algorithm

Answer: 

1. Depends only on locality in a canonical ordering - Lexicographic ordering works well for the web

2. Adjacency queries can be answered very efficiently - To fetch out-neighbors, trace back the chain of prototypes. This chain is typically short in practice

2. State two issues with the HITS algorithm.

Answer: 

1. Topic Drift - Off-topic pages can cause off-topic “authorities” to be returned. E.g., the neighborhood graph can be about a “super-topic”

2. Mutually Reinforcing Affiliates - Affiliated pages/sites can boost each other’s scores.
